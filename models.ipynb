```python
df_new = df_new[['num_io_operations', 'num_function_calls', 'var_declar','boolean_var','if_stm','method_invo','average_log_level',
                   'has_throw_statement','num_tokens', 'text_length', 'word_similarity', 'num_loop_iterations', 'num_catch_clauses',
                   'num_try_clauses', 'log_level']]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(['info', 'debug', 'warn', 'trace', 'error'])

df_new['log_level_ct'] = le.transform(df_new['log_level'])
```

```python
df_new.head()
```

```python
df_new = df_new.drop(['log_level'], axis = 1)
```

```python
# Assuming you have a 'TargetLabel' column in your dataset with binary labels (0 or 1)
target_labels = df_new["log_level_ct"]

# Remove the 'CodeSnippet' and 'TargetLabel' columns from your features DataFrame
X = df_new.drop(["log_level_ct"], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, target_labels, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn import svm, metrics, datasets

for i in range(8):
  X_train, X_test, y_train, y_test = train_test_split(X, target_labels, test_size=0.2)

  model1 = LogisticRegression(solver='sag', max_iter=10000)
  model2 = RandomForestClassifier(n_estimators=100)
  model3 = SVC(kernel='linear', C=1.0)
  model4 = AdaBoostClassifier(n_estimators=100)
  model5 = DecisionTreeClassifier(max_depth=None, min_samples_split=2, random_state=0)
  model6 = ExtraTreesClassifier(n_estimators=10, max_depth=None, min_samples_split=2, random_state=0)

  model1.fit(X_train, y_train)
  model1_pred = model1.predict(X_test)
  print("LR: {}".format(accuracy_score(y_test, model1_pred)))
  print("Classification report for - \n{}:\n{}\n".format(
    model1, metrics.classification_report(y_test, model1_pred)))

  model2.fit(X_train, y_train)
  model2_pred = model2.predict(X_test)
  print("RF: {}".format(accuracy_score(y_test, model2_pred)))
  print("Classification report for - \n{}:\n{}\n".format(
    model2, metrics.classification_report(y_test, model2_pred)))

  # model3.fit(X_train, y_train)
  # model3_pred = model3.predict(X_test)
  # print("SVM: {}".format(accuracy_score(y_test, model3_pred)))

  model4.fit(X_train, y_train)
  model4_pred = model4.predict(X_test)
  print("AdaB: {}".format(accuracy_score(y_test, model4_pred)))
  print("Classification report for - \n{}:\n{}\n".format(
    model4, metrics.classification_report(y_test, model4_pred)))

  model5.fit(X_train, y_train)
  model5_pred = model5.predict(X_test)
  print("DT: {}".format(accuracy_score(y_test, model5_pred)))
  print("Classification report for - \n{}:\n{}\n".format(
    model5, metrics.classification_report(y_test, model5_pred)))

  model6.fit(X_train, y_train)
  model6_pred = model6.predict(X_test)
  print("ExT: {}".format(accuracy_score(y_test, model6_pred)))
  print("Classification report for - \n{}:\n{}\n".format(
    model6, metrics.classification_report(y_test, model6_pred)))

  print("-----------------------------------------------------")
```
